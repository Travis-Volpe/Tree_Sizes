```python
def new_program (row):
#    if row['Class_Code'] == [other]:
#        return 'Other'
    if row['Class_Code'] == 119:
        return 'Tree Planting / CTP'
    if row['Class_Code'] == 120:
        return 'Tree Planting / CTP'
    if row['Class_Code'] == 125:
        return 'Riversmart'
    if row['Class_Code'] == 130:
        return 'Riversmart'
    if row['Class_Code'] == 155:
        return 'Riversmart'
    if row['Class_Code'] == 132:
        return 'Large Parcel'
    if row['Class_Code'] == 136:
        return 'Large Parcel'
    if row['Class_Code'] == 139:
        return 'Large Parcel'
    if row['Class_Code'] == 140:
        return 'Large Parcel'
    if row['Class_Code'] == 138:
        return 'Sustainable DC'
    if row['Class_Code'] == 135:
        return 'Sustainable DC'
    if row['Class_Code'] == 119.0:
        return 'Tree Planting / CTP'
    if row['Class_Code'] == 120.0:
        return 'Tree Planting / CTP'
    if row['Class_Code'] == 125.0:
        return 'Riversmart'
    if row['Class_Code'] == 130.0:
        return 'Riversmart'
    if row['Class_Code'] == 155.0:
        return 'Riversmart'
    if row['Class_Code'] == 132.0:
        return 'Large Parcel'
    if row['Class_Code'] == 136.0:
        return 'Large Parcel'
    if row['Class_Code'] == 139.0:
        return 'Large Parcel'
    if row['Class_Code'] == 140.0:
        return 'Large Parcel'
    if row['Class_Code'] == 138.0:
        return 'Sustainable DC'
    if row['Class_Code'] == 135.0:
        return 'Sustainable DC'
    return "Other"

dbtree_sizes['new_program'] = dbtree_sizes.apply(lambda row: new_program(row), axis=1 )

dbtree_sizes
```

```python
grouping2b = dbtree_sizes.groupby(['new_program'])
size_sum2b = grouping2b.describe()
size_sum2b
```
